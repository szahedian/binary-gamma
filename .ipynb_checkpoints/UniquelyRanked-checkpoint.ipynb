# Uniquely Ranked and Uniquely Weighted

In this notebook, we verify that the profiles used in the proof of Proposition 2.13 are uniquely ranked (see Definition 3.12) and uniquely weighted. 

```python
from pref_voting.profiles import *
from pref_voting.voting_methods import *
```

```python
def level_to_string(level): 
    l_str = str(level)
    if l_str[-1] == "1": 
        return l_str + "st"
    elif l_str[-1] == "2": 
        return l_str + "nd"
    elif l_str[-1] == "3": 
        return l_str + "rd"
    else:
        return l_str + "th"

def is_uniquely_ranked(prof): 
    """ 
    Check if a profile is uniquely ranked as defined in Definition 2.12.
    """
    for cands_to_remove in powerset(prof.candidates): 
        if len(cands_to_remove) < len(prof.candidates) - 1:
            reduced_prof, cmap = prof.remove_candidates(list(cands_to_remove))
            
            borda_scores = reduced_prof.borda_scores()
            
            b_scores = list(borda_scores.values())
            if len(b_scores) != len(list(set(b_scores))): 
                print("Profile is not uniquely ranked: \n")
                print("The Borda scores are not unique when restricted to the candidates: ", 
                      ', '.join([prof.cmap[cmap[c]] for c in reduced_prof.candidates]))
                reduced_prof.display(cmap={c: prof.cmap[cmap[c]] for c in reduced_prof.candidates})
                for c, bs in reduced_prof.borda_scores().items():
                    print(f"The Borda score of {prof.cmap[cmap[c]] } is {bs}.")
                return False
            num_levels = len(reduced_prof.candidates)
            for level in range(1, num_levels+1): 
                nranks = [reduced_prof.num_rank(cand, level) for cand in reduced_prof.candidates]
                if len(nranks) != len(list(set(nranks))): 
                    print("Profile is not uniquely ranked: \n")
                    print(f"The candidates do not have the same have the same number of {level_to_string(level)} place votes when restricted to the candidates: ", 
                      ', '.join([prof.cmap[c] for c in reduced_prof.candidates]))
                    reduced_prof.display(cmap={c: prof.cmap[cmap[c]] for c in reduced_prof.candidates})
                    for c in reduced_prof.candidates: 
                        print(f"{prof.cmap[cmap[c]]} is ranked {level_to_string(level)} by {reduced_prof.num_rank(c, level)} voters.")
                    return False
    return True


def uniquely_ranked_report(prof): 
    """ 
    For each subset of candidates, display the Borda scores for each candidate and the number of voters
    that rank the candidates in each possibility position. 
    """
    for cands_to_remove in powerset(prof.candidates): 
        if len(cands_to_remove) < len(prof.candidates) - 1:
            reduced_prof, cmap = prof.remove_candidates(list(cands_to_remove))
            print("Restricting to the candidates: ", 
                  ', '.join([prof.cmap[cmap[c]] for c in reduced_prof.candidates]))
            
            reduced_prof.display(cmap={c: prof.cmap[cmap[c]] for c in reduced_prof.candidates})
            print()
            borda_scores = reduced_prof.borda_scores()
            
            for c, bs in reduced_prof.borda_scores().items():
                print(f"\tThe Borda score of {prof.cmap[cmap[c]] } is {bs}.")

            print()
            num_levels = len(reduced_prof.candidates)
            for level in range(1, num_levels+1): 
                
                for c in reduced_prof.candidates: 
                    print(f"\t{prof.cmap[cmap[c]]} is ranked {level_to_string(level)} by {reduced_prof.num_rank(c, level)} voters.")
                print()

            print()
            
        
```

## $\mathbf{P}_0$

```python

a=0
b=1
c=2

p_0 = Profile([
    [a, b, c],
    [b, c, a],
    [c, a, b],
    [c, b, a]
],
    [44, 12, 36, 4],
    cmap = {a:"a", b:"b", c:"c"}

)


p_0.display()
print()

if is_uniquely_ranked(p_0): 
    print("The profile is uniquely ranked.")
else: 
    print("The profile is not uniquely ranked.")
print("\n")

if p_0.is_uniquely_weighted(): 
    print("The profile is uniquely weighted.")
else: 
    print("The profile is not uniquely ranked.")

print("\n\nTo check that the profile is uniquely weighted and uniquely ranked: \n\n")
print("The margin graph is: ")
p_0.display_margin_graph()

print()
uniquely_ranked_report(p_0)
```
